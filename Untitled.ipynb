```python
import numpy as np
import matplotlib.pyplot as plt
# generating signal:
samplingFrequency = 10*10**6 # hertz
signalFrequency = 1*10**3 # hertz
signalDuration = 0.01 # seconds
time = np.linspace(0, signalDuration, int(samplingFrequency*signalDuration))
signal = np.sin(2*np.pi*signalFrequency*time)
noise = np.random.normal(0, 1, size=np.size(signal))
signal = signal + noise
plt.plot(time, signal)
plt.show()
```

```python
signalF = np.fft.fftshift(np.fft.fft(signal))
sF_dB = 20*np.log10(np.abs(signalF))
sF_dB = sF_dB[int(np.size(sF_dB)/2):]
freq = np.linspace(0, int(samplingFrequency/2), np.size(sF_dB))
plt.plot(freq, sF_dB)
plt.grid()
plt.show()
print(freq[np.argmax(sF_dB)])
```
