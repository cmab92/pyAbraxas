```python
import numpy as np
import matplotlib.pyplot as plt
# generating signal:
samplingFrequency = 10*10^3 # hertz
signalFrequency = 1*10^3 # hertz
signalDuration = 0.5 # seconds
time = np.linspace(0, signalDuration, int(samplingFrequency*signalDuration))
signal = np.sin(2*np.pi*signalFrequency*time)
signal = signal + np.random.rand(np.size(signal))
plt.plot(time, signal)
plt.show()
```

```python
signalF = np.fft.fftshift(np.fft.fft(signal))[int(np.size(signalF)/2):]
freq = np.linspace(0, int(samplingFrequency/2), np.size(signalF))
plt.plot(freq, signalF)
```
